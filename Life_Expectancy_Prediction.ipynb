```python
y=data["Life expectancy"]
X=data.drop(["Life expectancy"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Veri sızıntısını (Data Leakage) önlemek için, eksik verileri doldurmadan önce veri setini Train ve Test olarak ayırıyorum.
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
# Önce ülkelere göre gruplama yaparak enterpolasyon uyguluyorum.
# Enterpolasyon ile dolmayan değerleri ise gelişmişlik statüsüne göre gruplayıp ortalama ile dolduruyorum.
```

```python
nan_olan_sutunlar=["Hepatitis B","Total expenditure","GDP","Population","Income composition of resources","Schooling","Alcohol"]

for sutun in nan_olan_sutunlar:
    X_train[sutun] = X_train.groupby("Country")[sutun].transform(lambda x: x.interpolate())
    X_train[sutun] = X_train.groupby("Status_Developing")[sutun].transform(lambda x: x.fillna(x.mean()))
    
    X_test[sutun] = X_test.groupby("Country")[sutun].transform(lambda x: x.interpolate())
    X_test[sutun] = X_test.groupby("Status_Developing")[sutun].transform(lambda x: x.fillna(X_train[sutun].mean()))

# Multicollinearity yaratmaması ve modelin kafasını karıştırmaması için birbirini tekrar eden, yüksek korelasyonlu sütunları siliyorum.
silinecekler = ["percentage expenditure", "thinness 5-9 years", "infant deaths", "Country"]
X_train.drop(silinecekler, axis=1, inplace=True)
X_test.drop(silinecekler, axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
y_pred_test=regression.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
score=print("Score:",r2_score(y_test,y_pred_test))
```

```python
from sklearn.linear_model import Lasso
regression=Lasso()
regression.fit(X_train,y_train)
y_pred_test=regression.predict(X_test)
from sklearn.metrics import r2_score
score=print("Score:",r2_score(y_test,y_pred_test))
```

```python
from sklearn.linear_model import Ridge
regression=Ridge()
regression.fit(X_train,y_train)
y_pred_test=regression.predict(X_test)
from sklearn.metrics import r2_score
score=print("Score:",r2_score(y_test,y_pred_test))
```

```python
from sklearn.linear_model import ElasticNet
regression=ElasticNet()
regression.fit(X_train,y_train)
y_pred_test=regression.predict(X_test)
from sklearn.metrics import r2_score
score=print("Score:",r2_score(y_test,y_pred_test))
```

```python
from sklearn.linear_model import LassoCV
regression=LassoCV(cv=5)
regression.fit(X_train,y_train)
y_pred_test=regression.predict(X_test)
from sklearn.metrics import r2_score
score=print("Score:",r2_score(y_test,y_pred_test))
```

```python
from sklearn.linear_model import RidgeCV
regression = RidgeCV(cv = 5)
regression.fit(X_train,y_train)
y_pred_test = regression.predict(X_test)
from sklearn.metrics import r2_score
score=print("Score:" ,r2_score(y_test,y_pred_test))
```

```python

from sklearn.linear_model import ElasticNetCV
regression = ElasticNetCV(cv = 5)
regression.fit(X_train,y_train)
y_pred_test = regression.predict(X_test)
from sklearn.metrics import r2_score

score=print("Score:" ,r2_score(y_test,y_pred_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# Decision Tree modelinin ezberlemesini (overfitting) engellemek ve en optimum ağaç yapısını bulmak için GridSearchCV kullandım.
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={
"max_depth":[3,5,10],
"min_samples_split":[2, 5, 10, 20],
"min_samples_leaf":[1,5,10,20]
}
```

```python
grid=GridSearchCV(estimator=DecisionTreeRegressor(),param_grid=param_grid,cv=5)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

```python
print("Score:",r2_score(y_test,y_pred))
print("mse=",mean_squared_error(y_test,y_pred))
print("mae=",mean_absolute_error(y_test,y_pred))
```
